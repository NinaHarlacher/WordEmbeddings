# Word2Vec using Genism 

```python
from keras.preprocessing.text import text_to_word_sequence

from multiprocessing import cpu_count
import pprint

from gensim.models.word2vec import Word2Vec
from gensim.models import KeyedVectors
import gensim.downloader as api
import gensim

from glob import glob
```

## Word2Vec Models

### Using a Pretrained Model

```python
pretrained = api.load('word2vec-google-news-300')
pretrained.save('pretrained.model')
```

```python
pretrained = KeyedVectors.load("pretrained.model")
```

#### Train a Model yourself

```python
#list of all texts in foler called corpus 
corpus = glob("corpus/*txt*") 
#preprocessing of data
data = [text_to_word_sequence(open(text).read(), lower=False ) for text in corpus]
#training and saving the model
trained_model=Word2Vec(data, size=300, window=2, min_count=1, sg=1,negative=5)
trained_model.save('trained.model')
```

```python
trained_model = KeyedVectors.load("trained.model")
```

## Most similar Word Embeddings



```python
 
list_words=[
"Shakespeare", 
"Hamlet",
"Ophelia",
   
"king",
"queen",
"prince",
"princess",
   
"ghost",
"angel",
"god",
"devil",
"witch",
"monster",
"water",
"cleanse",
"good",
"bad",
"gay",
"crazy",
"hysterical",
"die",
"love",
"heavy",
"marry",
"twixt",
"troth",
"wow",
"prithee",
"peace",
"war",
   
"France",
"Spain",
"Denmark",
"America", 
   
"cat",
"dog",
"fish",
"snake",
    
"Christian",
"Catholic",
"Jew",
"Protestant", 
"Muslim"
]
```

```python
dict_similar={} 
for word in list_words:
    dict_similar[word] = {}
    dict_similar[word]["Plays"] = []
    dict_similar[word]["GNews"] = []
    try: 
        dict_similar[word]["Plays"] = [similar[0] for similar in trained_model.wv.most_similar(word)]
    except:
        dict_similar[word]["Plays"].append("Word not contained in Elisabethan corpus")

        
    try:
        dict_similar[word]["GNews"] = [similar[0] for similar in pretrained.wv.most_similar(word)]
    except:
        dict_similar[word]["GNews"].append(["Word not contained in Google news corpus"] )       
    
    
```

## Vector Arithemtic 

```python
word_pairs=[
['woman', 'king', 'man'],
['man', 'queen', 'woman'],
['princess', 'woman', 'prince'],
    
["bigger" ,"small","big"],
["fastest", "slow", "fast"],
["prettier", "ugly", "pretty"],

["done", "make", "do"],
["seen", "say", "see"],
    
["France", "Denmark", "Paris"],
    
["cats", "dog", "cat"], 
]   

    
    

```

```python
dict_result={}
for words in word_pairs:
    key= words[0], "+", words[1], "-", words[2]
    dict_result[key]={}
    dict_result[key]["Plays"]= 0
    dict_result[key]["GNews"]= 0 
    dict_result[key]["Plays"]= [simi[0] for simi in trained_model.wv.most_similar(positive=[words[0], words[1]], 
                                                                                  negative=[words[2]], topn=3)]
    dict_result[key]["GNews"]= [simi[0] for simi in pretrained.wv.most_similar(positive=[words[0], words[1]], 
                                                                               negative=[words[2]], topn=3)]
```

```python
pprint.pprint(dict_result)
```
